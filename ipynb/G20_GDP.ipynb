# G20 국가별 GDP 크롤링, 데이터 저장

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('C:/ChromeDriver_exe/chromedriver_102.exe')
```

```python
# driver = webdriver.Chrome('C:/chromedriver.exe')
```

```python
url = 'https://ko.tradingeconomics.com/country-list/gdp-per-capita?continent=g20'
driver.get(url)
html = driver.page_source
```

```python
soup = bs(html, 'html.parser')
```

```python
th = soup.select('div.table-responsive tr th')
th
```

```python
columus = []
for i in range(len(th)) :
    columus.append(th[i].text)
columus
```

```python
td = soup.select('div.table-responsive tbody tr td')
td
```

```python
df = pd.DataFrame(columns=['국가','이전','현재','단위'])
contry_list =[]
before_list =[]
after_list =[]
unit_list =[]

for i in range(0,len(td),5):
    contry_list.append(td[i].text.strip())
    before_list.append(td[i+2].text.strip())
    after_list.append(td[i+1].text.strip())
    unit_list.append(td[i+4].text.strip())
df['국가'] = contry_list
df['이전'] = before_list
df['현재'] = after_list
df['단위'] = unit_list
```

```python
df
```

```python
df.to_csv('./g20_gdp.csv', index = False)
```
