# 국내 서빙로봇 도입 규모
- 업계추정치
- 출처 : 서울경제(www.sedaily.com)

```python
data3 = pd.DataFrame()
data3['연도'] = ['2019','2020','2021','2022']
data3['서빙로봇규모'] = [50,500,2100,3000]
data3
```

```python
data3.to_csv('../data/연도별국내서빙로봇도입규모.csv', index = False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(20,10))

fig.set_facecolor('#EEEEEE')
plt.gca().set_facecolor('#EEEEEE')
plt.rcParams['lines.linewidth'] = 5

plt.rc('font', family = 'Malgun Gothic', size = 18)

plt.plot(data3['연도'], data3['서빙로봇규모'], color = '#F2B117', 
         marker = 'o', ms = 15)

plt.title('국내 서빙로봇 도입 규모(업계추정치)', 
          fontsize = 20,loc='center', pad=30, fontweight="bold")
plt.ylabel('서빙로봇 규모')

plt.text(-0.15,3550, '단위: 대', fontsize=15, color = 'black')
plt.ylim(0,3500)

plt.xticks([0,1,2,3], data3['연도']+'년')

# # 그래프 상단에 값 추가하기
for i in range(len(data3['서빙로봇규모'])):
    height = data3['서빙로봇규모'][i:i+1]
    plt.text(data3['연도'][i:i+1], height + 10, '%.1i' %height, 
             ha='center', va='bottom', size = 40, color='black')

plt.show()
```
